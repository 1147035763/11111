```python
import pandas as pd
import pymysql
db_info = {
    'host':'127.0.0.1',
    'user':'root',
    'password':'123',
    'database':'ll',
    'charset':'utf8'
}


day = pymysql.connect(**db_info)

result_day = 'select * from day'

day = pd.read_sql(result_day,day)

day
```

```python
import pandas as pd
import jieba
from snownlp import SnowNLP 
import pyecharts.options as opts
from pyecharts.charts import WordCloud
from pyecharts.globals import SymbolType

day['词性标注'] = day['评论内容'].agg(lambda x:list(SnowNLP(x).tags))

list_tags = []
for i in day['词性标注']:
    for j in i:
        list_tags.append(j)
        
def heat_count(name,series):
    count = 0
    for i in series:
        for j in i:
            if j==name:
                count+=1
    return count

day_tags = pd.DataFrame(list_tags,columns=['词语','词性'])

day_tags['词性'].value_counts()
result = day_tags[day_tags['词性']=='a'].groupby(by='词语')['词语'].count().sort_values(ascending=False)[:1000]
day_wordcloud_tags = [(i,str(j)) for i,j in zip(result.index,result.values)]

c = (
    WordCloud()
    .add(series_name="", 
         data_pair=day_wordcloud_tags,
         word_size_range=[30,100],
         textstyle_opts=opts.TextStyleOpts(font_family="华文彩云"),
        )
    .set_global_opts(
        title_opts=opts.TitleOpts(
            title="B站视频评论", title_textstyle_opts=opts.TextStyleOpts(font_size=23)
        ),
        tooltip_opts=opts.TooltipOpts(is_show=True),
    )
)
c.render_notebook()
```

# 数据库表格2

```python
import pandas as pd
import pymysql
db_info = {
    'host':'127.0.0.1',
    'user':'root',
    'password':'123',
    'database':'ll',
    'charset':'utf8'
}


day2 = pymysql.connect(**db_info)

result_day2 = 'select * from day2'

day2 = pd.read_sql(result_day2,day2)

day2
```

```python
import pandas as pd
from pyecharts import options as opts
from pyecharts.charts import Bar, Line
from pyecharts import options as opts
from pyecharts.charts import Radar,Grid,Bar,Timeline
from pyecharts.globals import ThemeType

data1=day2.loc[day2["名字"]=="佩恩",["最高击杀","最高死亡","最高助攻"]].values.tolist()
data2=day2.loc[day2["名字"]=="佩恩",["场均击杀","场均死亡","场均助攻"]].values.tolist()
day_list=day2["名字"][::-1]
from pyecharts import options as opts
from pyecharts.charts import Radar,Grid,Bar,Timeline
from pyecharts.globals import ThemeType

def abc(year):
    b=(
    Bar()
    .add_xaxis(["最高击杀","最高死亡","最高助攻"])
    .add_yaxis("能力值",day2.loc[day2["名字"]==year,["最高击杀","最高死亡","最高助攻"]].values.tolist()[0])
    .set_global_opts(legend_opts=opts.LegendOpts(pos_left="70%",pos_top="0%"))
    )
    b2=(
    Bar()
    .add_xaxis(["场均击杀","场均死亡","场均助攻"])
    .add_yaxis("场均能力",day2.loc[day2["名字"]==year,["场均击杀","场均死亡","场均助攻"]].values.tolist()[0])
    .reversal_axis()
    .set_series_opts(label_opts=opts.LabelOpts(position="right"))
    .set_global_opts(legend_opts=opts.LegendOpts(pos_left="25%",pos_top="0%"))
    )
    grid_chart=(
            Grid()
        .add(
                b,
                grid_opts=opts.GridOpts(pos_left="55%",pos_top="10%")
        )
        .add(
                b2,
                grid_opts=opts.GridOpts(pos_right="55%",pos_top="10%")
    )
    )
    return grid_chart
timeline=Timeline(
    init_opts=opts.InitOpts(width="1000px",height="600px"
                           )
)

timeline.add_schema(
    orient="vertical",
    is_auto_play=True,
    is_inverse=True,
    play_interval=2000,
    pos_left="null",
    pos_right="0%",
    pos_top="20",
    width="50",
    label_opts=opts.LabelOpts(is_show=True,position='left'),
)
for day in day_list:
    day_grid_chart=abc(day)
    timeline.add(day_grid_chart,day)
timeline.render_notebook()
```

```python
from pyecharts import options as opts
from pyecharts.charts import Pie
from pyecharts.faker import Faker
import pandas as pd


def age_range(age):
    if age<2:
        return '0-2'
    elif age<3:
        return '2-3'
    elif age<5:
        return '3-5'
    elif age<8:
        return '5-8'
    else:
        return '8以上'
day2['段'] = day2['最高死亡'].agg(age_range)
result = day2.groupby(by='段')['最高死亡'].count()
result
for x,y in zip(['a','b'], [1,2]):
    print(x) # 输出元组对象（和列表类似的对象）
    print(y)
list = [[x,int(y)] for x,y in zip(result.index, result.values)]
list   
l = (
    Pie(init_opts=opts.InitOpts(width="600px", height="300px")) # 设置背景的大小
    .add(
        series_name = "段", # 必须项
        data_pair = list,
        radius=["25%", "45%"], # 设置环的大小
#         center=["20%", "50%"], # 设置饼图的位置
        rosetype="radius", # 设置玫瑰图类型
        label_opts=opts.LabelOpts(
            position="outside",
            formatter="{type|{b}的个数: }{number|{c}} {bg1|}\n {percent|占比：{d}%}{bg2|} ", # 定义节点名称 type\number\percent
            background_color="#fff", # 大标签的背景颜色
            border_color="black",
            border_width=1,
            border_radius=1,
            rich={
                "type": {"fontSize": 12, "lineHeight": 10, "align": "center"},#字体调小
                "number": {"fontSize": 12, "lineHeight": 10, "align": "center","backgroundColor": "yellow"},#字体调小
                "bg1":{"height": 22}, # 控制第一行的高度
                "percent": {
                    "fontSize": 12,
                    "color": "white", # 字的颜色
                    "backgroundColor": "#aaa", # 模块背景颜色
                    "padding": [5, 20], # 控制模块的高度和长短
                    "borderRadius": 5, # 控制模块的角的形状
                    "height": 10, # 标题框高度
                    "align": "center",
                },
                "bg2":{"height": 20,},# 控制第二行的高度
            },
        ),
    )
    .set_colors(["pink", "blue", "orange", "red"]) # 颜色设置
    .set_global_opts(title_opts=opts.TitleOpts(title="最高死亡玫瑰图"),
                    legend_opts=opts.LegendOpts(pos_top="10%", pos_left="86%"), # 设置图示的位置
                    )
)
l.render_notebook()
```

# beautifulsoup库（bs4)修改html文件

```python
from bs4 import BeautifulSoup
```

```python
with open("page.html", "r+", encoding='utf-8') as html:
    html_bf = BeautifulSoup(html, 'lxml')
    divs = html_bf.select('.chart-container') # 根据css定位标签，选中图像的父节点标签
    divs[0]["style"] = "width:420px;height:55px;position:absolute;top:20px;left:460px;border-style:dashed;border-color:#AAAAAA;border-width:0px;"
    divs[1]["style"] = "width:900px;height:400px;position:absolute;top:100px;left:60px;border-style;border-color:#FFFFF0;border-width:2px;"
    divs[2]["style"] = "width:750px;height:450px;position:absolute;top:500px;left:60px;border-style;border-color:#FFFFF0;border-width:2px;"
    divs[3]["style"] = "width:700px;height:450px;position:absolute;top:500px;left:530px;border-style;border-color:#FFFFF0;border-width:2px;"
    divs[4]["style"] = "width:750px;height:450px;position:absolute;top:500px;left:1100px;border-style;border-color:#FFFFF0;border-width:2px;"
    divs[5]["style"] = "width:900px;height:400px;position:absolute;top:100px;left:900px;border-style;border-color:#AAAAAA;border-width:2px;"
    body = html_bf.find("body") # 根据标签名称定位到body标签
    body["style"] = "background-color:#AAAAAA;" # 修改背景颜色
    html_new = str(html_bf) # 将BeautifulSoup对象转换为字符
    html.seek(0, 0) # 光标移动至
    html.truncate() # 删除光标后的所有字符内容
    html.write(html_new) # 将由BeautifulSoup对象转换得到的字符重新写入html文件
    html.close()
```
